```python
import re
import requests

def get_cve_numbers(url):
    response = requests.get(url)
    html = response.text
    pattern =  re.compile(r'CVE-\d{4}-\d{4,7}')
    cve_numbers = re.findall(pattern, html)
    return cve_numbers

url = "https://cve.mitre.org/data/downloads/allitems-cvrf-year-2023.xml"
cve_numbers = get_cve_numbers(url)

if cve_numbers:
    print("Found the following CVE numbers:")
    for cve_number in cve_numbers:
        print(cve_number)
else:
    print("No CVE numbers found.")
```
